```python
import numpy as np

X_train = np.load("../processed/X_train.npy")   # (N, 150, 6)
X_test   = np.load("../processed/X_test.npy")
y_train = np.load("../processed/y_train.npy")
y_test   = np.load("../processed/y_test.npy")

num_classes = len(np.unique(y_train))
```

```python
from tensorflow.keras.models import load_model

cnn = "../model/air_gesture_cnn.h5"
lstm = "../model/air_gesture_lstm.h5"
cnn_lstm = "../model/air_gesture_cnn_lstm.h5"

cnn_model = load_model(cnn)
lstm_model = load_model(lstm)
cnn_lstm_model = load_model(cnn_lstm)
```

```python
from sklearn.metrics import classification_report, f1_score
import numpy as np

# Prediction
y_pred_prob = cnn_model.predict(X_test)
y_pred = np.argmax(y_pred_prob, axis=1)

# True labels (already correct)
y_true = y_test

# Sanity check
print(y_true.shape, y_pred.shape)

# Metrics
print(classification_report(y_true, y_pred))
print("Weighted F1-score:", f1_score(y_true, y_pred, average="weighted"))
```
